```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
Shopping=pd.read_csv("online_shoppers_intention.csv")
```

```python
Shopping.describe(include='all')
```

```python
#Get the Correlation values for all features
print(Shopping.corr())
```

```python
#Create histogram for distribution of values in Target variable
shopping_target_plain = Shopping['Revenue']
class_dist = shopping_target_plain.value_counts()
plt.bar(class_dist.index, class_dist)
plt.ylabel("Frequency")
plt.show()
```

```python
#Function to generate scatter plot for outliers
def scatter_graph(x,y):
    plt.scatter(Shopping[x], Shopping[y])
# show the plot
    plt.show()
```

```python
scatter_graph('ProductRelated_Duration','ExitRates')
```

```python
scatter_graph('ProductRelated_Duration','Administrative_Duration')
```

```python
scatter_graph('ProductRelated_Duration','BounceRates')
```

```python
#Function to generate box plots for outliers
def plot_graph(col):
    sns.boxplot(x=Shopping[col])
```

```python
plot_graph('Administrative_Duration')
```

```python
plot_graph('Administrative')
```
